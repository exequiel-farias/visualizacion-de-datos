# Test de hipotesis

```python
sns.lmplot(x='iu',y='valor_factura',data=tarjetas, col='moroso');
```

```python
from scipy.stats import ranksums
```

Ahora realizaremos un test de hipótesis con la variable moroso, para conocer si estos dos grupos (moroso y no moroso) son estadísticamente diferentes o no.

H<sup>null<sup>

La distribución de los grupos moroso y no moroso es la misma

H<sup>alt<sup>

La distribución de los grupos moroso y no moroso no es la misma

```python
#Crear grupos moroso y no moroso
moroso = tarjetas.query("moroso == 1").valor_factura
no_moroso = tarjetas.query("moroso == 0").valor_factura
```

```python
moroso
```

```python
no_moroso
```

```python
#Vemos si estos dos grupos si la diferencia en la distribucion es significativa
resultado = ranksums(moroso, no_moroso)
```

```python
print("El p.value de nuestro test es {}".format(resultado.pvalue))
```

Podemos ver que el pvalue está debajo del 5% por lo que sirve para aceptar la hipotesis alternativa. Entonces podemos decir la distribución de los grupos moroso y no moroso no es la misma
